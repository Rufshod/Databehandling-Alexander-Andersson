b)
- Combine year and week number

```python
df.insert(0, "Week", df["år"].astype(str) + "v" + df["veckonummer"].astype(str)) # Combining År and Veckonummer into Vecka. Added a v in between
df.pop("år") # Removing år and veckonummer
df.pop("veckonummer")

df["Week"].head()
```

```python
df.describe()
```

```python
# Translating columns into english:
df.rename(
    columns={
        "Antal_avlidna_vecka": "Amount dead per week",
        "Antal_fall_vecka": "Amount of cases per week",
        "Kum_antal_fall": "Cumulative amount of cases",
    },
    inplace=True,
)
```

c)- f) seaborn

```python

year = df["Week"]

fig, ax = plt.subplots(2,2, figsize = (16, 12), dpi = 120)
sns.set_style("whitegrid")
sns.lineplot(df, x = year, y = df["Amount dead per week"], ax = ax[0,0])
sns.lineplot(df, x = year, y = df["Amount of cases per week"], ax = ax[0,1])

sns.lineplot(df, x = year, y = df["Amount dead per week"], ax = ax[1,0],label= "Deaths", linestyle="--")
sns.lineplot(df, x = year, y = df["Amount of cases per week"], ax = ax[1,0], label= "Cases").set_yscale("log") 

sns.lineplot(df, x = year, y = df["Cumulative amount of cases"], ax = ax[1,1]).set_yscale("log") 


ax[0,0].xaxis.set_major_locator(plt.MaxNLocator(9))
ax[0,1].xaxis.set_major_locator(plt.MaxNLocator(9))
ax[1,0].xaxis.set_major_locator(plt.MaxNLocator(9))
ax[1,1].xaxis.set_major_locator(plt.MaxNLocator(9))

ax[0,0].set(title="Amount dead per week")
ax[0,1].set(title="Amount of cases per week")
ax[1,0].set(title="Amount dead / Amount of cases per week")
ax[1,1].set(title="Cumulative amount of cases")
fig.suptitle("Covid cases 2020 week 6 to now.")
fig.tight_layout()
sns.set_style("whitegrid");


# Fix this with a loop. and add: .y_label("Amount of People") to all.
```
